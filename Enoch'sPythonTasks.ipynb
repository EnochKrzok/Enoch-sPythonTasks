````python
import random
import time

def random_array(size):
    array = [random.randint(0,9) for i in range(size)]
    return array

def swap(array, index_1, index_2):
  temp = array[index_1]
  array[index_1] = array[index_2]
  array[index_2] = temp

def bubble_sort_unoptimized(array):
  count = 0
  for x in array:
    for index in range(len(array) - 1):
      count += 1
      if array[index] > array[index + 1]:
        swap(array, index, index + 1) 
  print("Pre-optimized bubble_sort \nCount: {0}" .format(count))
  return array

def bubble_sort_optimized(array):
  count = 0
  for i in range(len(array)):
    for idx in range(len(array) - i - 1):
      count += 1
      if array[idx] > array[idx + 1]:
        array[idx], array[idx + 1] = array[idx + 1], array[idx]  
  print("Post-optimized bubble_sort \nCount: {0}" .format(count))
  return array

def merge_sort(array):
  if len(array) <= 1:
    return array
  middle_index = len(array) // 2
  left_split = array[:middle_index]
  right_split = array[middle_index:]
  left_sorted = merge_sort(left_split)
  right_sorted = merge_sort(right_split)
  return merge(left_sorted, right_sorted)

def merge(left, right):
  result = []
  while (left and right):
    if left[0] < right[0]:
      result.append(left[0])
      left.pop(0)
    else:
      result.append(right[0])
      right.pop(0)
  if left:
    result += left
  if right:
    result += right
  return result

def linear_search(array, find):
  for i in range(len(array)):
    if array[i] == find:
      return find, 'has been found at', i
  return find, 'not in list'

def binary_search(array, find):
  left_pointer = 0
  right_pointer = len(array)
  while left_pointer < right_pointer:
    mid_idx = (left_pointer + right_pointer) // 2
    mid_val = array[mid_idx]
    if mid_val == find:
      return find, 'value found at',mid_idx
    if find < mid_val:
      right_pointer = mid_idx
    if find > mid_val:
      left_pointer = mid_idx + 1
  return find, 'Value not in list'

array = random_array(10)
sorted_array = merge_sort(array) 
ran = random.randint(0,9)
print(array)
print(bubble_sort_unoptimized(array))
print(bubble_sort_optimized(array))
print('linear search\n', linear_search(array,ran))
print('binary search\n', binary_search(sorted_array,ran),'\n')

array2 = random_array(10000)
print(len(array2))
start = time.time()
bubble_sort_optimized(array2)
end = time.time()
total = end - start
print('it took', total, 'seconds\n')

array3 = random_array(10000)
start = time.time()
bubble_sort_optimized(array3)
end = time.time()
total1 = end - start
print('it took', total1, 'seconds')
start = time.time()
bubble_sort_unoptimized(array3)
end = time.time()
total2 = end - start
print('it took', total2, 'seconds')
print('the optimized sort was ', total2 - total1, 'seconds faster\n')

array4 = random_array(10000)
start = time.time()
bubble_sort_optimized(array4)
end = time.time()
total1 = end - start
print('it took', total1, 'seconds')
start = time.time()
bubble_sort_unoptimized(array4)
end = time.time()
total2 = end - start
print('it took', total2, 'seconds')
print('the optimized bubble sort was ', total2 - total1, 'seconds faster than unoptimized bubble sort')
start = time.time()
merge_sort(array4)
end = time.time()
total3 = end - start
print('it took', total3, 'seconds for merge sort')
print('the merge sort was ', total1 - total3, 'seconds faster than the optimized bubble sort')
print('the merge sort was ', total2 - total3, 'seconds faster than the unoptimized bubble sort\n')

sorted_array2 = merge_sort(array4)
sorted_array2[len(sorted_array2)-1]=999
start = time.time()
print(linear_search(sorted_array2,999))
end = time.time()
total1 = end - start
print('it took', total1, 'seconds')
start = time.time()
print(binary_search(sorted_array2,999))
end = time.time()
total2 = end - start
print('it took', total2, 'seconds')
print('the binary search was ', total1 - total2, ' seconds faster than linear search' )
````
